```python
import pytse_client as tse
import pandas as pd
```

```python
tickers = tse.download(symbols=["کارا","افران","اعتماد","کیان","یاقوت","سپیدما"],include_jdate=True, write_to_csv=True)
```

```python
df=pd.DataFrame()
df['Date']=tickers["اعتماد"]['jdate']
df['Afran']=tickers["افران"]['adjClose']
df['Kara']=tickers["کارا"]['adjClose']
df['Etemad']=tickers["اعتماد"]['adjClose']
df['Kian']=tickers["کیان"]['adjClose']
df['Yagut']=tickers["یاقوت"]['adjClose']
df['Sepidma']=tickers["سپیدما"]['adjClose']
```
